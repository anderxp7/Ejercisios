# CONJUNTOS

Un conjunto es una colección de objetos o elementos que comparten características comunes. Estos elementos, que pueden ser números, letras o cualquier otra cosa, se agrupan y el conjunto se denota con una letra mayúscula. Los conjuntos se pueden representar gráficamente mediante diagramas de Venn (usando círculos o elipses) y se definen mediante extensión (listando los elementos) o por comprensión (describiendo una propiedad común)

### Conjuntos en Python

Un conjunto es una colección desordenada de elementos únicos. En Python, se definen usando llaves `{}` o la función `set()`.

```python
# Creación de conjuntos
conjunto_a = {1, 2, 3, 4, 5}
conjunto_b = {4, 5, 6, 7, 8}
conjunto_c = set([5, 6, 9, 10])

print(f"Conjunto A: {conjunto_a}")
print(f"Conjunto B: {conjunto_b}")
print(f"Conjunto C: {conjunto_c}")

# Añadir elementos (los duplicados no se añaden)
conjunto_a.add(6)
conjunto_a.add(1)
print(f"Conjunto A después de añadir: {conjunto_a}")

# Eliminar elementos
conjunto_a.discard(1) # No lanza error si el elemento no existe
print(f"Conjunto A después de descartar 1: {conjunto_a}")

# Operaciones de conjuntos
union = conjunto_a.union(conjunto_b)
print(f"Unión de A y B: {union}")

interseccion = conjunto_a.intersection(conjunto_b)
print(f"Intersección de A y B: {interseccion}")

diferencia_ab = conjunto_a.difference(conjunto_b)
print(f"Diferencia A - B: {diferencia_ab}")

diferencia_ba = conjunto_b.difference(conjunto_a)
print(f"Diferencia B - A: {diferencia_ba}")

diferencia_simetrica = conjunto_a.symmetric_difference(conjunto_b)
print(f"Diferencia simétrica de A y B: {diferencia_simetrica}")
```

### Relaciones entre Conjuntos

Las relaciones entre conjuntos describen los vínculos entre ellos, como la inclusión (un conjunto está dentro de otro), la igualdad (comparten los mismos elementos), la intersección (comparten elementos en común) y la disyunción (no tienen elementos en común). Matemáticamente, una relación se define formalmente como un subconjunto del producto cartesiano de dos conjuntos.



 Relacion1.png

```python
conjunto_x = {1, 2, 3}
conjunto_y = {1, 2, 3, 4, 5}
conjunto_z = {6, 7}

print(f"Conjunto X: {conjunto_x}")
print(f"Conjunto Y: {conjunto_y}")
print(f"Conjunto Z: {conjunto_z}")

# Subconjunto (issubset)
es_subconjunto = conjunto_x.issubset(conjunto_y)
print(f"¿X es subconjunto de Y? {es_subconjunto}")

es_subconjunto_propio = conjunto_x < conjunto_y # Subconjunto propio (no son iguales)
print(f"¿X es subconjunto propio de Y? {es_subconjunto_propio}")

# Superconjunto (issuperset)
es_superconjunto = conjunto_y.issuperset(conjunto_x)
print(f"¿Y es superconjunto de X? {es_superconjunto}")

es_superconjunto_propio = conjunto_y > conjunto_x # Superconjunto propio (no son iguales)
print(f"¿Y es superconjunto propio de X? {es_superconjunto_propio}")

# Disjuntos (isdisjoint) - No tienen elementos en común
es_disjunto_xz = conjunto_x.isdisjoint(conjunto_z)
print(f"¿X y Z son disjuntos? {es_disjunto_xz}")

es_disjunto_xy = conjunto_x.isdisjoint(conjunto_y)
print(f"¿X y Y son disjuntos? {es_disjunto_xy}")

# Igualdad de conjuntos
conjunto_igual_x = {3, 1, 2}
print(f"¿X es igual a {{3, 1, 2}}? {conjunto_x == conjunto_igual_x}")
```

# Proposiciones

Una proposición es un enunciado declarativo con un valor de verdad bien definido . Las proposiciones son los componentes básicos de la lógica. Este capítulo presenta brevemente los fundamentos de la lógica proposicional. Se introducen los operadores lógicos básicos: negación, disyunción y conjunción.



desktop_079c7783-dcd9-43b7-a302-b9a93332b6ea.png

```python
# Definición de proposiciones (variables booleanas)
esta_lloviendo = True
hace_frio = False
tengo_paraguas = True

print(f"Está lloviendo: {esta_lloviendo}")
print(f"Hace frío: {hace_frio}")
print(f"Tengo paraguas: {tengo_paraguas}")

# Proposición simple (Negación)
no_esta_lloviendo = not esta_lloviendo
print(f"No está lloviendo: {no_esta_lloviendo}")

# Proposición compuesta (Conjunción - 'AND')
puedo_salir_con_paraguas = esta_lloviendo and tengo_paraguas
print(f"¿Puedo salir si está lloviendo y tengo paraguas?: {puedo_salir_con_paraguas}")

# Proposición compuesta (Disyunción - 'OR')
necesito_abrigo = hace_frio or esta_lloviendo
print(f"¿Necesito un abrigo si hace frío o está lloviendo?: {necesito_abrigo}")

# Proposición más compleja
puedo_salir_comodo = (not esta_lloviendo and not hace_frio) or (esta_lloviendo and tengo_paraguas)
print(f"¿Puedo salir cómodo (no llueve y no hace frío, o si llueve tengo paraguas)?: {puedo_salir_comodo}")

# Ejemplo de tabla de verdad con una función
def implicacion(p, q):
    # p -> q es equivalente a not p or q
    return not p or q

print("\nTabla de verdad para la implicación (p -> q):")
print(f"p=True, q=True: {implicacion(True, True)}")
print(f"p=True, q=False: {implicacion(True, False)}")
print(f"p=False, q=True: {implicacion(False, True)}")
print(f"p=False, q=False: {implicacion(False, False)}")
```

# Tablas de verdad

Las tablas de verdad son herramientas gráficas que muestran sistemáticamente todos los posibles valores de verdad de una proposición compuesta, dados los valores de verdad de sus proposiciones simples.



tablas-de-verdad.jpg

```python
print("Tabla de Verdad para AND (Conjunción):")
print("P\tQ\tP AND Q")
for p in [True, False]:
    for q in [True, False]:
        print(f"{str(p):<5}\t{str(q):<5}\t{str(p and q):<5}")

print("\nTabla de Verdad para OR (Disyunción):")
print("P\tQ\tP OR Q")
for p in [True, False]:
    for q in [True, False]:
        print(f"{str(p):<5}\t{str(q):<5}\t{str(p or q):<5}")

print("\nTabla de Verdad para NOT (Negación):")
print("P\tNOT P")
for p in [True, False]:
    print(f"{str(p):<5}\t{str(not p):<5}")

print("\nTabla de Verdad para Implicación (P -> Q):")
# P -> Q es equivalente a (not P) or Q
print("P\tQ\tP -> Q")
for p in [True, False]:
    for q in [True, False]:
        print(f"{str(p):<5}\t{str(q):<5}\t{str(not p or q):<5}")

print("\nTabla de Verdad para Bicondicional (P <-> Q):")
# P <-> Q es equivalente a (P -> Q) AND (Q -> P)
# O también (P and Q) or (not P and not Q)
print("P\tQ\tP <-> Q")
for p in [True, False]:
    for q in [True, False]:
        print(f"{str(p):<5}\t{str(q):<5}\t{str((p and q) or (not p and not q)):<5}")
```

# Cuantificadores

Los cuantificadores son símbolos utilizados en lógica matemática (específicamente en la lógica de predicados) para especificar la cantidad de elementos en un conjunto que cumplen con una determinada propiedad. Permiten transformar una función proposicional (un enunciado con variables abiertas que no es ni verdadero ni falso por sí mismo, como "x es un número par") en una proposición lógica con un valor de verdad definido (verdadero o falso).

```python
# Cuantificador Universal (Para todo - \u2200)
# En Python, podemos simularlo con la función `all()`

# Ejemplo 1: Todos los números en una lista son pares
numeros1 = [2, 4, 6, 8, 10]
todos_son_pares = all(num % 2 == 0 for num in numeros1)
print(f"Lista 1: {numeros1}")
print(f"\u00BFSon todos los números pares? {todos_son_pares}")

# Ejemplo 2: No todos los números en una lista son mayores que 5
numeros2 = [6, 7, 3, 9]
todos_mayores_que_5 = all(num > 5 for num in numeros2)
print(f"Lista 2: {numeros2}")
print(f"\u00BFSon todos los números mayores que 5? {todos_mayores_que_5}")

# Cuantificador Existencial (Existe al menos uno - \u2203)
# En Python, podemos simularlo con la función `any()`

# Ejemplo 3: Existe al menos un número impar en una lista
numeros3 = [2, 4, 5, 8, 10]
existe_impar = any(num % 2 != 0 for num in numeros3)
print(f"Lista 3: {numeros3}")
print(f"\u00BFExiste algón número impar? {existe_impar}")

# Ejemplo 4: No existe ningón número negativo en una lista
numeros4 = [1, 2, 3, 4]
existe_negativo = any(num < 0 for num in numeros4)
print(f"Lista 4: {numeros4}")
print(f"\u00BFExiste algôn número negativo? {existe_negativo}")

# Combinando cuantificadores con otras condiciones
# Ejemplo 5: Todas las personas en una lista son mayores de edad Y están activas
personas = [
    {"nombre": "Ana", "edad": 25, "activa": True},
    {"nombre": "Luis", "edad": 17, "activa": True},
    {"nombre": "Marta", "edad": 30, "activa": False}
]

todas_mayores_y_activas = all(p["edad"] >= 18 and p["activa"] for p in personas)
print(f"\nPersonas: {personas}")
print(f"\u00BFSon todas las personas mayores de edad Y activas? {todas_mayores_y_activas}")

# Ejemplo 6: Existe al menos una persona que sea menor de edad O no esté activa
existe_menor_o_inactivo = any(p["edad"] < 18 or not p["activa"] for p in personas)
print(f"\u00BFExiste al menos una persona que sea menor de edad O no esté activa? {existe_menor_o_inactivo}")
```

# Funciones

Una función en matemáticas es una regla o ley de correspondencia que asigna a cada elemento de un conjunto de entrada exactamente un elemento de un conjunto de salida. Son esenciales para describir cómo una cantidad depende de otra, por ejemplo, cómo la distancia recorrida depende del tiempo y la velocidad. Conceptos Clave de las Funciones Regla de Correspondencia: Es la operación o fórmula (como \(f(x)=2x\)) que define cómo los valores de entrada se transforman en valores de salida.

Variable Independiente x: Es la variable de entrada, cuyos valores provienen del conjunto inicial (dominio). Estos valores se eligen libremente.

Variable Dependiente( y ) Es la variable de salida, cuyo valor depende del valor de la variable independiente. Es el resultado de aplicar la regla de la función.

```python
# Definición de una función simple
def saludar(nombre):
    """Esta función saluda a la persona cuyo nombre se pasa como argumento.
    """
    return f"¡Hola, {nombre}! Bienvenido/a."

# Llamada a la función
mensaje1 = saludar("Ana")
print(mensaje1)

mensaje2 = saludar("Pedro")
print(mensaje2)

# Otra función que realiza una operación matemática
def sumar(a, b):
    """Esta función toma dos números y devuelve su suma.
    """
    resultado = a + b
    return resultado

# Llamada a la función de suma
suma_numeros = sumar(10, 5)
print(f"La suma es: {suma_numeros}")

suma_otros_numeros = sumar(-3, 7)
print(f"La suma es: {suma_otros_numeros}")
```

# Principio de induccion

La inducción matemática es un método de demostración utilizado para probar que una proposición, propiedad o fórmula es válida para todos los números naturales (o para un conjunto infinito de valores enteros a partir de un valor inicial). Se basa en el principio del buen orden y permite demostrar una infinidad de casos en un número finito de pasos.

Principio y Analogía
El principio de inducción matemática se puede comparar con la idea de una fila interminable de fichas de dominó: si se demuestra que la primera ficha cae, y que la caída de cualquier ficha hace que caiga la siguiente, se puede concluir que todas las fichas caerán.

### Ilustración del Principio de Inducción Matemática en Python

Vamos a ilustrar el principio de inducción matemática utilizando la proposición de que la suma de los primeros $n$ números naturales es igual a $n(n+1)/2$. Es decir, $P(n): 1 + 2 + \dots + n = \frac{n(n+1)}{2}$.

**Recordatorio del Principio:**
1.  **Caso Base ($P(1)$)**: Demostrar que la proposición es cierta para $n=1$.
2.  **Paso Inductivo ($P(k) \implies P(k+1)$)**: Asumir que la proposición es cierta para algún entero positivo $k$ (Hipótesis de Inducción), y luego demostrar que es cierta para $k+1$.

```python
# Función para calcular la suma de los primeros n números naturales
def sum_n_natural_numbers(n):
    return sum(range(1, n + 1))

# Función para calcular la fórmula propuesta
def formula_sum_n_natural_numbers(n):
    return n * (n + 1) / 2

print("--- Caso Base (n=1) ---")
n = 1
sum_actual = sum_n_natural_numbers(n)
sum_formula = formula_sum_n_natural_numbers(n)

print(f"Suma real para n={n}: {sum_actual}")
print(f"Suma segón fórmula para n={n}: {sum_formula}")

if sum_actual == sum_formula:
    print(f"El caso base P({n}) es cierto.")
else:
    print(f"El caso base P({n}) NO es cierto.")
```

### Paso Inductivo ($P(k) \implies P(k+1)$)

En Python, no podemos 'demostrar' simbólicamente el paso inductivo, pero podemos ilustrarlo mostrando que si la hipótesis de inducción ($P(k)$ es cierta) se cumple, entonces $P(k+1)$ también parece cumplirse. Esto se hace mostrando que:

$1 + 2 + \dots + k + (k+1) = \frac{(k+1)((k+1)+1)}{2}$

Si $P(k)$ es cierto, entonces $1 + 2 + \dots + k = \frac{k(k+1)}{2}$. Sustituyendo esto en el lado izquierdo de la ecuación de $P(k+1)$:

$\frac{k(k+1)}{2} + (k+1) = (k+1)(\frac{k}{2} + 1) = (k+1)(\frac{k+2}{2}) = \frac{(k+1)(k+2)}{2}$

Que es exactamente el lado derecho de la fórmula para $P(k+1)$.

Podemos verificar esto numéricamente para algunos valores de $k$.

```python
print("--- Ilustración del Paso Inductivo ---")

# Verificamos para algunos valores de k
for k in range(1, 5):
    print(f"\nVerificando para k={k} (asumiendo P(k) es cierto):")

    # Lado izquierdo de P(k+1) usando la Hipótesis de Inducción
    # Suma de 1 a k + (k+1)
    lhs_pk_plus_1 = formula_sum_n_natural_numbers(k) + (k + 1)
    print(f"LI de P(k+1) (vía Hipótesis): {lhs_pk_plus_1}")

    # Lado derecho de P(k+1) con la fórmula directamente para (k+1)
    rhs_pk_plus_1 = formula_sum_n_natural_numbers(k + 1)
    print(f"LD de P(k+1) (vía Fórmula): {rhs_pk_plus_1}")

    if lhs_pk_plus_1 == rhs_pk_plus_1:
        print(f"Para k={k}, si P({k}) es cierto, P({k+1}) también lo es.")
    else:
        print(f"Para k={k}, algo no es correcto en el paso inductivo.")
```

# Funcion recursiva

es una función que se llama a sí misma dentro de su propia definición para resolver un problema, dividiéndolo en subproblemas más pequeños hasta que se llega a un caso base que no requiere más llamadas. Este proceso requiere un caso base para detener las llamadas recursivas y así evitar una ejecución infinita.

Este proceso requiere un caso base para detener las llamadas recursivas y así evitar una ejecución infinita.

Caso base: Es la condición que detiene la recursión. Sin un caso base, la función se llamaría a sí misma indefinidamente

```python
def factorial(n):
    # Caso base: cuando n llegue a 0, devolvemos 1
    if n == 0:
        return 1

    # Llamada recursiva
    return n * factorial(n - 1)

# Ejemplo de uso
numero = 5
print("El factorial de", numero, "es:", factorial(numero))

def sumatoria(n):
    # Caso base: la suma hasta 1 es 1
    if n == 1:
        return 1

    # Llamada recursiva
    return n + sumatoria(n - 1)

# Ejemplo de uso
numero = 6
print("La sumatoria hasta", numero, "es:", sumatoria(numero))

```

#Teoria de grafos


Los grafos se utilizan para representar redes de transporte, conexiones sociales, circuitos eléctricos, rutas de comunicación, mapas, procesos lógicos, relaciones entre datos y, en general, cualquier situación donde existan objetos que interactúan. Su versatilidad hace que sean una herramienta fundamental en ingeniería, ciencias de la computación, matemáticas aplicadas, inteligencia artificial, optimización y análisis de redes.

Dentro de los grafos existen diferentes tipos según sus características: dirigidos o no dirigidos, ponderados o no ponderados, simples o multigrafos, y modelos más complejos como árboles, grafos bipartitos o grafos completos. El análisis de estas estructuras permite resolver problemas importantes, como encontrar caminos más cortos, detectar ciclos, optimizar rutas, buscar conexiones mínimas, identificar componentes críticas o modelar sistemas inteligentes.

Caminosmascortos.jpg



```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Creación de un Grafo
print("--- Creación y Visualización de un Grafo ---")
G = nx.Graph() # Grafo no dirigido
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5), (5, 6), (6, 3)])

# Visualizar el grafo
plt.figure(figsize=(8, 6))
nx.draw_networkx(G, with_labels=True, node_color='skyblue', node_size=1500, font_size=10)
plt.title("Grafo de Ejemplo")
plt.show()


# 2. Caminos (Paths)
print("\n--- Caminos ---")
node_start = 1
node_end = 5

if nx.has_path(G, node_start, node_end):
    print(f"Existe un camino entre el nodo {node_start} y el nodo {node_end}.")
    # Encontrar un camino simple (sin nodos repetidos)
    path = nx.shortest_path(G, source=node_start, target=node_end)
    print(f"Camino más corto de {node_start} a {node_end}: {path}")
else:
    print(f"No existe camino entre el nodo {node_start} y el nodo {node_end}.")


# 3. Ciclos (Cycles)
print("\n--- Ciclos ---")
if nx.is_directed(G):
    # Para grafos dirigidos, se busca circuit_basis
    cycles = list(nx.simple_cycles(G))
else:
    # Para grafos no dirigidos, se busca cycle_basis
    cycles = list(nx.cycle_basis(G))

if cycles:
    print("Se encontraron los siguientes ciclos en el grafo:")
    for i, cycle in enumerate(cycles):
        print(f"  Ciclo {i+1}: {cycle}")
else:
    print("No se encontraron ciclos en el grafo.")


# 4. Conectividad
print("\n--- Conectividad ---")
if nx.is_connected(G):
    print("El grafo es CONEXO (todos los nodos están conectados entre sí).")
else:
    print("El grafo NO es CONEXO.")
    components = list(nx.connected_components(G))
    print(f"Componentes conexas: {components}")


# 5. Árboles (Trees)
# Un árbol es un grafo conexo y sin ciclos.
print("\n--- Árboles ---")

# Crear un grafo que sea un árbol
T = nx.Graph()
T.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5)])

plt.figure(figsize=(8, 6))
nx.draw_networkx(T, with_labels=True, node_color='lightgreen', node_size=1500, font_size=10)
plt.title("Grafo Árbol de Ejemplo")
plt.show()

print("\nVerificando si el Grafo T es un árbol:")
is_T_connected = nx.is_connected(T)
has_T_cycles = nx.cycle_basis(T)

if is_T_connected and not has_T_cycles:
    print(f"El grafo T es un ÁRBOL (es conexo y no tiene ciclos).")
else:
    print(f"El grafo T NO es un árbol. Conexo: {is_T_connected}, Tiene ciclos: {bool(has_T_cycles)}.")

print("\nVerificando si el Grafo G original es un árbol:")
is_G_connected = nx.is_connected(G)
has_G_cycles = nx.cycle_basis(G)

if is_G_connected and not has_G_cycles:
    print(f"El grafo G es un ÁRBOL (es conexo y no tiene ciclos).")
else:
    print(f"El grafo G NO es un árbol. Conexo: {is_G_connected}, Tiene ciclos: {bool(has_G_cycles)}.")
```

# Introduccion algebra booleana


Introducción al Álgebra Booleana

El álgebra booleana es una rama de las matemáticas que estudia sistemas lógicos basados en dos valores:
0 y 1, también interpretados como Falso y Verdadero, Bajo y Alto, Off y On.

Fue desarrollada por George Boole en el siglo XIX y constituye la base teórica de los circuitos digitales, los computadores, los microcontroladores y en general de toda la electrónica digital moderna.

En el álgebra booleana los valores se combinan mediante operaciones lógicas, entre las cuales las más importantes son:

AND ( ∧ ) → Conjunción

OR ( ∨ ) → Disyunción

NOT ( ¬ ) → Negación

Además, existen otras operaciones derivadas como XOR, NAND y NOR, ampliamente utilizadas en diseño de circuitos.

Una característica fundamental del álgebra booleana es que sus variables y operaciones siguen leyes y propiedades específicas, como la ley conmutativa, distributiva, de identidad, de complemento y de absorción, que permiten simplificar expresiones lógicas y optimizar circuitos digitales.

El álgebra booleana permite representar y simplificar funciones lógicas que después pueden implementarse mediante puertas lógicas o mediante programación en sistemas digitales como Arduino, ESP32, PLC, FPGAs, entre otros.
```python
# --- Ejemplos de Álgebra Booleana en Python ---

# Definición de las operaciones booleanas básicas
def boolean_and(a, b):
    return a and b

def boolean_or(a, b):
    return a or b

def boolean_not(a):
    return not a

print("---------------------------------------------------")
print("Ejemplo 1: Evaluación de una condición lógica compuesta")
print("---------------------------------------------------")

# Variables booleanas
es_dia = True
hace_sol = False
tengo_paraguas = True

# Condición lógica 1: ¿Puedo salir sin paraguas si es de día Y hace sol?
# (es_dia AND hace_sol) AND (NOT tengo_paraguas)

resultado_and = boolean_and(es_dia, hace_sol)
resultado_not_paraguas = boolean_not(tengo_paraguas)
condicion_final_1 = boolean_and(resultado_and, resultado_not_paraguas)

print(f"Es de día: {es_dia}")
print(f"Hace sol: {hace_sol}")
print(f"Tengo paraguas: {tengo_paraguas}")
print(f"¿Es de día Y hace sol? {resultado_and}")
print(f"¿NO tengo paraguas? {resultado_not_paraguas}")
print(f"Condición final: ¿Puedo salir sin paraguas si es de día Y hace sol?: {condicion_final_1}")

print("\n---------------------------------------------------")
print("Ejemplo 2: Simulación simple de una puerta lógica (AND y OR)")
print("---------------------------------------------------")

# Simulación de una puerta AND
input_a = True
input_b = True
output_and = boolean_and(input_a, input_b)
print(f"Entrada A: {input_a}, Entrada B: {input_b} -> Salida AND: {output_and}")

input_a = True
input_b = False
output_and = boolean_and(input_a, input_b)
print(f"Entrada A: {input_a}, Entrada B: {input_b} -> Salida AND: {output_and}")

# Simulación de una puerta OR
input_c = False
input_d = True
output_or = boolean_or(input_c, input_d)
print(f"Entrada C: {input_c}, Entrada D: {input_d} -> Salida OR: {output_or}")

input_c = False
input_d = False
output_or = boolean_or(input_c, input_d)
print(f"Entrada C: {input_c}, Entrada D: {input_d} -> Salida OR: {output_or}")
```
